```python
# Import library
from sklearn.model_selection import train_test_split
# Putting feature variables into X
X = data.drop(['Class'], axis=1)
# Putting target variable to y
y = data['Class']
# Splitting data into train and test set 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=100)
```

```python
# Standardization method
from sklearn.preprocessing import StandardScaler
# Instantiate the Scaler
scaler = StandardScaler()
# Fit the data into scaler and transform
X_train['Amount'] = scaler.fit_transform(X_train[['Amount']])
X_train.head()
```

```python
# Transform the test set
X_test['Amount'] = scaler.transform(X_test[['Amount']])
X_test.head()
```

```python
# Importing metrics
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, f1_score
from sklearn.metrics import accuracy_score
```

```python
results = pd.DataFrame(columns=['Model Name', 'Accuracy', 'F1-score'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
# Instantiate the model with best C
logistic = LogisticRegression(C=0.01)
# Fit the model on the train set
logistic_model = logistic.fit(X_train, y_train)
# Create an empty DataFrame
results = pd.DataFrame(columns=['Model Name', 'Accuracy', 'F1-score'])
# Prepare results function
def display_test_results(model_name, model):
    # Prediction on the test set
    y_test_pred = model.predict(X_test)
    # Calculate F1-score and accuracy
    f1 = f1_score(y_test, y_test_pred)
    accuracy = accuracy_score(y_test, y_test_pred)
    # Add results to the DataFrame
    results.loc[len(results)] = [model_name, accuracy, f1]
    return None
# Call the function to display and add results
display_test_results("Logistic Regression", logistic_model)
print(results)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Instantiate the Random Forest model
random_forest = RandomForestClassifier(n_estimators=100, random_state=0)  # You can adjust the hyperparameters as needed
# Fit the model on the train set
random_forest_model = random_forest.fit(X_train, y_train)
# Prepare results function
def display_test_results(model_name, model):
    # Prediction on the test set
    y_test_pred = model.predict(X_test)
    # Calculate F1-score and accuracy
    f1 = f1_score(y_test, y_test_pred)
    accuracy = accuracy_score(y_test, y_test_pred)
    # Add results to the DataFrame
    results.loc[len(results)] = [model_name, accuracy, f1]
    return None
# Call the function to display and add results for Random Forest
display_test_results("Random Forest", random_forest_model)
# Print the results DataFrame
print(results)
```

```python
from sklearn.svm import SVC
# Instantiate the SVM model
svm = SVC(kernel='linear', C=1.0)  # You can adjust the kernel and hyperparameters as needed
# Fit the model on the train set
svm_model = svm.fit(X_train, y_train)
# Prepare results function
def display_test_results(model_name, model):
    # Prediction on the test set
    y_test_pred = model.predict(X_test)
    # Calculate F1-score and accuracy
    f1 = f1_score(y_test, y_test_pred)
    accuracy = accuracy_score(y_test, y_test_pred)
    # Add results to the DataFrame
    results.loc[len(results)] = [model_name, accuracy, f1]
    return None
# Call the function to display and add results for SVM
display_test_results("SVM", svm_model)
# Print the results DataFrame
print(results)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Instantiate the Decision Tree model
decision_tree = DecisionTreeClassifier(random_state=0)  # You can adjust the hyperparameters as needed
# Fit the model on the train set
decision_tree_model = decision_tree.fit(X_train, y_train)
# Prepare results function
def display_test_results(model_name, model):
    # Prediction on the test set
    y_test_pred = model.predict(X_test)
    # Calculate F1-score and accuracy
    f1 = f1_score(y_test, y_test_pred)
    accuracy = accuracy_score(y_test, y_test_pred)
    # Add results to the DataFrame
    results.loc[len(results)] = [model_name, accuracy, f1]
    return None
# Call the function to display and add results for Decision Tree
display_test_results("Decision Tree", decision_tree_model)
# Print the results DataFrame
print(results)
```
